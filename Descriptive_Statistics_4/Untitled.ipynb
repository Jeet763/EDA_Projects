```python
# Topic 3: Handling Missing Values

data3 = {
    'customer_id': [1001, 1002, 1003, 1004, 1005, 1006],
    'income': [550000, None, 1200000, 'NA', 980000, 920000],
    'loan_amount': [500000, 800000, 1500000, None, 1200000, 'XX'],
    'credit_score': [720, 680, None, 650, 740, 999]
}
df_missing = pd.DataFrame(data3)
df_missing
```

```python
# STEP 1: Convert indicators like 'NA', 'XX', '999' to np.nan

df_missing = df_missing.replace(['NA', 'XX', 'NaN'], np.nan)
print("After converting 'NA', 'XX', '999' to np.nan:")
print(df_missing)
```

```python
# STEP 2: Drop rows if income is missing (assuming a few only are missing)
df_dropped = df_missing.dropna(subset=['income'])
print("After dropping rows where income is missing:")
print(df_dropped)
```

```python
# STEP 3: Impute missing loan_amount using mean (business judgement)

mean_loan = pd.to_numeric(df_dropped['loan_amount'] , errors = 'coerce').mean()
df_dropped['loan_amount'] = pd.to_numeric(df_dropped['loan_amount'], errors = 'coerce').fillna(mean_loan)
print(f"Imputed missing loan amount with mean ({mean_loan}):")
df_dropped
```

```python
# STEP 4: Impute missing credit_score with median

# median_score = df_dropped['credit_score'].median()
# df_dropped['credit_score'] = df_dropped['credit_score'].fillna(median_score)
# print(f"Imputed missing credit_score with median ({median_score}):")
# df_dropped

median_score = df_dropped['credit_score'].median()

df_dropped['credit_score'] = df_dropped['credit_score'].fillna(median_score)

print(f"Imputed missing credit_score with median ({median_score}):")
df_dropped
```

```python
# Topic 4: Handling Outliers

# Simulate BFSI dataset with transaction outliers
data4 = {
    'customer_id': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    'transaction_amount': [50000, 120000, 20000, 90000, 220000, 10500000, 80000]
}

df_outliers = pd.DataFrame(data4)
df_outliers
```

```python
# STEP 1: Detect outliers using IQR

Q1 = df_outliers['transaction_amount'].quantile(0.25)
Q3 = df_outliers['transaction_amount'].quantile(0.75)
```

```python
IQR = Q3 - Q1
print(f"IQR method  Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```

```python
outliers = df_outliers[(df_outliers['transaction_amount'] < lower_bound) | 
                       (df_outliers['transaction_amount'] > upper_bound)]

print("Detected outliers:")
print(outliers)
```

```python
# STEP 2: Remove outliers (for analysis purposes)

df_no_outliers = df_outliers[(df_outliers['transaction_amount'] >= lower_bound) & 
                             (df_outliers['transaction_amount'] <= upper_bound)]

print(df_no_outliers)
```

```python
# STEP 3: Cap outliers (alternative to removal)

df_capped = df_outliers.copy()
df_capped.loc[df_capped['transaction_amount'] > upper_bound, 'transaction_amount'] = upper_bound
print("Transactions after capping outlier values to upper bound:")
print(df_capped)
```

```python
# Simulated balances in a BFSI context
data5 = pd.DataFrame({
    'customer_id': [1, 2, 3],
    'balance': ['2.5 lakh', '100000', '1.2 lakh']
})
data5
```

```python
# Function to convert 'lakh' to rupees

def convert_lakh(x):
    if 'lakh' in str(x):
        return float(x.replace('lakh' , '').strip()) * 100000
    else:
        return float(x)

data5['balance_rupees'] = data5['balance'].apply(convert_lakh)
data5
```

```python
data5['balance_thousands'] = data5['balance_rupees'] /1000
print(data5[['customer_id', 'balance_thousands']])
```

```python
data5['balance_standard'] = data5['balance_rupees'].round(2)
print(data5[['customer_id', 'balance_standard']])
```

```python
df = pd.DataFrame({'transaction_amount': [1000, 1200, 1150, 1100, 50000]})
```

```python
df_std = df['transaction_amount'].std()
df_std
```

```python
df['z_score'] = (df['transaction_amount'] - df['transaction_amount'].mean()) / df['transaction_amount'].std()
df
```

```python
# Mark as outlier if |z_score| > 3
df['is_outlier'] = df['z_score'].abs() > 3
print(df)
```

```python
Q1 = df['transaction_amount'].quantile(0.25)
Q3 = df['transaction_amount'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df['is_outlier_iqr'] = (df['transaction_amount'] < lower) | (df['transaction_amount'] > upper)
print(df)
```
